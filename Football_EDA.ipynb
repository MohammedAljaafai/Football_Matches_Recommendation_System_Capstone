# **Football Mathces Datasets**

### **Data Description:**
- These datasets include the football matches results of the last 10 seasons plus the current season. 
- These results are for the English, German, Spanish, Italic, French, Portuguese, Dutch, and Saudi leagues

### **Data Dictionary:**
The English, German, Spanish, Italic, French, Portuguese, And Dutch leagues shares these 10 columns

Key to results data:
- League = Referese to the country's league
- Wk = The week of the game during the season
- Day= The Day of the game during the season
- Date = The Date of the football game
- Time = The Timing of the football game
- Home = Home team
- Home = Away team
- Score = Game Score (Number of goals for each team)
- Attendance = Total attendance of the game
- Venue = Stadium Name
- Referee = Referee Name
- Match Report = A report for the match (If there are important information. For examplem a team forfeit from a game)
- Notes = Notes

```python
#Identify common features for all datasets
Features = ['League', 'Wk', 'Day', 'Date', 'Time', 'Home', 'Score', 'Away',
       'Attendance', 'Venue', 'Referee', 'Match Report', 'Notes']
len(Features)
```

```python
#This function is for handelling the unwanted features
def my_function(g):
    c=[]
    for (columnName, columnData) in g.iteritems():
        z=columnName in Features
        if z == False:
            c.append(columnName)
    print(c)        
    g.drop(c, axis=1 , inplace = True)
```

```python
#Read Leagues datasets and mergethem in a dataset for each league
all_files0 = glob.glob('Data/Spanish' + "/*.csv")
spanish = pd.concat(map(pd.read_csv, all_files0), ignore_index=True, sort=False)
spanish.insert(0,'League', 'SP')
my_function(spanish)

all_files1 = glob.glob('Data/Portuguese' + "/*.csv")
portuguese = pd.concat(map(pd.read_csv, all_files1), ignore_index=True, sort=False)
portuguese.insert(0,'League', 'P')
my_function(portuguese)

all_files2 = glob.glob('Data/Italic' + "/*.csv")
italic = pd.concat(map(pd.read_csv, all_files2), ignore_index=True, sort=False )
italic.insert(0,'League', 'IT')
my_function(italic)

all_files3 = glob.glob('Data/French' + "/*.csv")
french = pd.concat(map(pd.read_csv, all_files3), ignore_index=True, sort=False)
french.insert(0,'League', 'FL')
my_function(french)

all_files4 = glob.glob('Data/German' + "/*.csv")
german = pd.concat(map(pd.read_csv, all_files4), ignore_index=True, sort=False)
german.insert(0,'League', 'BL')
my_function(german)

all_files5 = glob.glob('Data/English' + "/*.csv")
english = pd.concat(map(pd.read_csv, all_files5), ignore_index=True, sort=False)
english.insert(0,'League', 'PL')
my_function(english)

all_files6 = glob.glob('Data/Dutch' + "/*.csv")
dutch = pd.concat(map(pd.read_csv, all_files6), ignore_index=True, sort=False)
dutch.insert(0,'League', 'D')
my_function(dutch)

all_files7 = glob.glob('Data/Saudi' + "/*.csv")
saudi = pd.concat(map(pd.read_csv, all_files7), ignore_index=True, sort=False)
saudi.insert(0,'League', 'SPL')
my_function(saudi)
```

## Data Munging

```python
# Checking the most common Scores
df.Score.value_counts()
```

```python
# Defining new columns
df['Home_Team_Score']=np.nan
df['Away_Team_Score']=np.nan
```

```python
# This loop is used to identify the index of scores 10-2, 10-0 
for i in range(len(df)):
    if df['Score'][i] == '10–0':
        print(i)
    if df['Score'][i] == '10–2':
        print(i)
```

```python
#Creat Home team goals, and Away team goals Columns

for i in range(len(df)):
    if pd.isnull(df['Score'][i]) == False and i != 1674 and i != 6311:
        df['Home_Team_Score'][i] = df['Score'][i][0]
        df['Away_Team_Score'][i] = df['Score'][i][2]
    if i == 1674 or i == 6311:
        df['Home_Team_Score'][i] = df['Score'][i][0]+df['Score'][i][1]
        df['Away_Team_Score'][i] = df['Score'][i][3]    
```

```python
#Droppint the observations with missing values in the score feature
for i in range(len(df)):
    if pd.isnull(df['Score'][i]) == True:
       df=df.drop(index=i)
#Dropping observations 25164 and 26441, because the value of the score is not understandable
df.drop(index=25164,axis=0,inplace=True)
df.drop(index=26441, axis=0,inplace=True)
df.reset_index(drop= True, inplace=True)
```

```python
#Convert the features to its proper type
df['Wk'] = df['Wk'].astype(str)
for i in range(len(df)):
        if pd.isnull(df['Home_Team_Score'][i]) == False:
                df['Home_Team_Score'] = df['Home_Team_Score'].astype(int)
        if pd.isnull(df['Away_Team_Score'][i]) == False:
                df['Away_Team_Score'] = df['Away_Team_Score'].astype(int)
```

```python
df.info()
```

```python
#add new Columns, Total Number of goals, and score difference
df['Total_goals']= df['Home_Team_Score']+df['Away_Team_Score']
df['goals_diff']= abs(df['Home_Team_Score']-df['Away_Team_Score'])
```

```python
# adding a column for the game (EXAMPLE, team_x vs team_y)
df['Game']=''
for i in range(len(df)):
    df['Game'][i]= df['Home'][i] + ' vs ' + df['Away'][i]
```

```python
# adding a column for Home_wins, Away_Wins, Draws
df['Home_wins']=''
df['Away_Wins']=''
df['Draws']=''
for i in range(len(df)):
    if df.Home_Team_Score[i] > df.Away_Team_Score[i]:
        df['Home_wins'][i] = 'Yes'
        df['Away_Wins'][i] = 'No'
        df['Draws'][i]='No'
    if df.Home_Team_Score[i] < df.Away_Team_Score[i]:
        df['Home_wins'][i] = 'No'
        df['Away_Wins'][i] = 'Yes'
        df['Draws']='No' 
    if df.Home_Team_Score[i] - df.Away_Team_Score[i] ==0:
        df['Home_wins'][i] = 'No'
        df['Away_Wins'][i] = 'No'
        df['Draws'][i]='Yes'

df.sample(5)
```

```python
#Creating A dataframe for the avrage results of each game for all games 
All_games=df[['Game', 'Attendance', 'Total_goals', 'goals_diff']]
All_games = All_games.groupby('Game').mean().reset_index()
All_games['Total_goals'] = All_games['Total_goals'].astype(int)
All_games['goals_diff'] = All_games['goals_diff'].astype(int)
All_games['Game_Class']=''
All_games.sample(5)

for i in range(len(All_games)):
    
    if All_games.Total_goals[i] >= 3 and All_games.goals_diff[i] >= 3: #example 3-0 is a one side game or 4-1, 5-2, 6-3
        All_games['Game_Class'][i]='One_Side_Game'
    if (All_games.Total_goals[i] >= 2 and All_games.Total_goals[i] <= 3) and All_games.goals_diff[i] <= 2: #example 2-0, 2-1 are a Competitive games        
        All_games['Game_Class'][i]='Competitive_Game'
    if All_games.Total_goals[i] >= 4 and All_games.goals_diff[i] <= 2: #example 2-2, 3-1 4-2, 4-3, 5-3, 5-4, 6-4, 3-3, 4-4, 5-5, 6-6 are competitve open (Have a lot of goals) games
        All_games['Game_Class'][i]='Competitive_Open_Game'

    if All_games.Total_goals[i] <= 2 and All_games.goals_diff[i] <= 1 : #example 1-1 is a closed game or 1-0, 0-0, 1-1
        All_games['Game_Class'][i]='Closed_Game'
All_games.reset_index(drop= True, inplace=True)
```

```python
# Creating a new dataframe that contains the number of Home_wins, Away_Wins, Draws
Home_wins=0
Away_Wins=0
Draws=0
for i in range(len(df)):
    if df.Home_Team_Score[i] > df.Away_Team_Score[i]:
        Home_wins = Home_wins + 1
    if df.Home_Team_Score[i] < df.Away_Team_Score[i]:
        Away_Wins = Away_Wins + 1
    else:
        Draws = Draws + 1

Results=['Home_wins', 'Away_Wins', 'Draws']
Count= [Home_wins, Away_Wins, Draws]
results_count = pd.DataFrame()
results_count['Results'] = Results
results_count['Count'] = Count
results_count
```

```python
fig = px.bar(results_count.groupby(Results).Count.mean()
 ,title="Away Team Goals count")

fig.update_layout(
    xaxis_title="Number of Goals of the Away Team",
    yaxis_title="Count",
    barmode='stack', xaxis={'categoryorder':'total descending'}
    )

fig.show()
```

```python
#Creating a new dataframe that contains home, away, wins and draws, total wins ans losses and winning percentage
Teams_win_loss_count = pd.DataFrame()
Teams_win_loss_count['All_Teams'] = df.Home.unique()
Teams_win_loss_count['Home_wins'] = 0
Teams_win_loss_count['Home_losses'] = 0
Teams_win_loss_count['Away_wins'] = 0
Teams_win_loss_count['Away_losses'] = 0
Teams_win_loss_count['Home_Draws'] =  0
Teams_win_loss_count['Away_Draws'] =  0
Teams_win_loss_count.set_index("All_Teams", inplace = True)


for i in range(len(df)):
    if df.Home_Team_Score[i] > df.Away_Team_Score[i]:
        Teams_win_loss_count['Home_wins'][df.Home[i]]= Teams_win_loss_count['Home_wins'][df.Home[i]] + 1
        Teams_win_loss_count['Away_losses'][df.Home[i]]= Teams_win_loss_count['Away_losses'][df.Home[i]] + 1
    if df.Home_Team_Score[i] < df.Away_Team_Score[i]:
        Teams_win_loss_count['Away_wins'][df.Away[i]]= Teams_win_loss_count['Away_wins'][df.Away[i]] + 1
        Teams_win_loss_count['Home_losses'][df.Home[i]]= Teams_win_loss_count['Home_losses'][df.Home[i]] + 1
    if df.Home_Team_Score[i] == df.Away_Team_Score[i]:
        Teams_win_loss_count['Home_Draws'][df.Home[i]]= Teams_win_loss_count['Home_Draws'][df.Home[i]] + 1    
        Teams_win_loss_count['Away_Draws'][df.Away[i]]= Teams_win_loss_count['Away_Draws'][df.Away[i]] + 1
  
Teams_win_loss_count['Total_wins'] = Teams_win_loss_count['Home_wins'] + Teams_win_loss_count['Away_wins']
Teams_win_loss_count['Total_losses'] = Teams_win_loss_count['Home_losses'] + Teams_win_loss_count['Away_losses']
Teams_win_loss_count['Total_Draws'] = Teams_win_loss_count['Home_Draws'] + Teams_win_loss_count['Away_Draws']
Teams_win_loss_count['Winning_percentage'] = (Teams_win_loss_count['Total_wins'] / (Teams_win_loss_count['Total_losses'] + Teams_win_loss_count['Total_Draws'] + Teams_win_loss_count['Total_wins']))*100
```
