```python
All_games2 = pd.DataFrame()
def inputgame(d):
    global All_games2
    for i in range(len(d)):
        All_games2=  All_games2.append(All_games[All_games['Game'].str.contains(d[i])])
    return All_games2.reset_index(drop= True, inplace=True)
```

```python
inputgame(Tomorrow)
All_games=All_games2
```

```python
All_games['id'] = All_games.groupby(['Game']).ngroup()
All_games
```

```python
features = ['Game', 'Attendance', 'Total_goals', 'goals_diff', 'Game_Class', 'League']
for feature in features:
    All_games[feature] = All_games[feature].fillna('')
```

```python
All_games.Attendance = All_games.Attendance.astype(str)
All_games.Total_goals = All_games.Total_goals.astype(str)
All_games.goals_diff = All_games.goals_diff.astype(str)
```

```python
def combined_features(row):
    return row['Game_Class'] +" "+row['Attendance']+" "+row['Total_goals']+" "+row['goals_diff']+" "+row['League']
All_games["combined_features"] = All_games.apply(combined_features, axis =1)
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(All_games["combined_features"])
print("Count Matrix:", count_matrix.toarray())
```

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python
game_user_likes = "Al-Shabab vs Al-Ittihad"
def get_index_from_title(Game):
    return All_games[All_games.Game == Game]['id'].values[0]
game_index = get_index_from_title(game_user_likes)
```

```python
similar_games = list(enumerate(cosine_sim[game_index]))
```

```python
sorted_similar_games = sorted(similar_games, key=lambda x:x[1], reverse=True)
```
